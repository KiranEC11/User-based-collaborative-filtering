```python
import pandas as pd
import numpy as np
```

```python
df_ratings = pd.read_csv('ratings.csv')
df_ratings.head()
```

```python
rating_matrix = df_ratings.pivot_table(index='userId', columns='movieId', values='rating', fill_value=0)
rating_matrix
```

```python
## load movie dataset

df_movies = pd.read_csv('movies.csv')
df_movies.head()
```

```python
df_merged = pd.merge(df_movies,df_ratings, on='movieId')
df_merged
```

```python
## linear kernel

from sklearn.metrics.pairwise import linear_kernel
```

```python
## cosine similarity

cosine_sim = linear_kernel(rating_matrix,rating_matrix)
```

```python
np.argmax(cosine_sim[100])
```

```python
# user id

user_id = rating_matrix.index.values
```

```python
# movie id

movie_id = rating_matrix.columns.values
```

```python
def find_movie_id(col_idx):
    return movie_id[col_idx]
```

```python
def similar_users(user):
    sim_vector = cosine_sim[user]
    sorted_vector = sorted(list(enumerate(sim_vector)), key=lambda x:x[1], reverse=True)[1:11]  
    sim_users = [tup[0] for tup in sorted_vector]   ## top 10 similar users
    return sim_users
```

```python
## top movies rated by given user( rating=5)

def top_movies_rated(user):
    df = df_ratings.loc[df_ratings['userId']==user]
    df_sorted = df.sort_values(by = "rating", ascending =False)[['movieId','rating']]
    
    top_movies = df_sorted.loc[df_sorted['rating']==5]['movieId'].values
    return top_movies
    
```

```python
## movies watched by a given user

def movies_watched(user):
    df_user = df_ratings.loc[df_ratings['userId']==user]
    return df_user['movieId'].values
```

```python
def movie_title(movie_id):
    return df_merged.loc[df_merged['movieId']==movie_id]['title'].values[0]
```

```python
## recommend movie ids

def rec_movies(user):
    top=[]
    sim_users = similar_users(user)
    for user in sim_users:
        top.append(top_movies_rated(user))
    top = np.concatenate(top) ## convert the list of arrays into one dim array
    
    movie, freq = np.unique(top, return_counts=True)
    freq_sorted = sorted(list(enumerate(freq)), key=lambda x: x[1], reverse=True)
    top_movies = [movie[item[0]] for item in freq_sorted]  
    """
    top movies to recommend, 
    but we need to eliminate the movies that are already watched by given user.
    """
    user_watched_movies = movies_watched(user)
    movie_ids = [movie for movie in top_movies if movie not in user_watched_movies][:10]
    movies_recommend = [movie_title(movie_id) for movie_id in movie_ids ]
    
    return movies_recommend
```

```python
rec_movies(600)
```
